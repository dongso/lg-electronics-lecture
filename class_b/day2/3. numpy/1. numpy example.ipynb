```python
import matplotlib.pyplot as plt
import numpy as np
```

### 1. 계단 오르내리기 - Python

```python
%%time
position = 0
walk = [position]
steps = 1000

for i in range(1000):
    step = 1 if np.random.randint(0, 2) else -1
    position += step
    walk.append(position)
```

```python
plt.plot(walk)
```

### 2. 계단 오르내리기 - Numpy

```python
%%time
nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps)
steps = np.where(draws > 0, 1, - 1)
walk = steps.cumsum()
```

```python
plt.plot(walk)
```

### 3. 계단 오르내리기 - Numpy 5000번

```python
nwalks = 5000
nsteps = 1000
```

```python
draws = np.random.randint(0, 2, size=(nwalks, nsteps))
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(axis=1)
```
